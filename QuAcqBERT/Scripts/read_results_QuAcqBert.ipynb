```python
def showResults(prob,results):
  print("-------------------"+prob+":\n")

  # all converged
  print("all converged ?: ",all([r["TargetEquivalentToLearned"] for r in results]))
  # nbConverged
  print("nb converged ?: ",sum([1 for r in results if r["TargetEquivalentToLearned"]]))

  # size of basis
  sizesOfBasis = [r["sizeOfBasis"] for r in results]
  average2 = sum(sizesOfBasis)/len(sizesOfBasis)

  print("Average size of Basis: ",average2)

  # |T|
  sizesOfTarget = [r["sizeOfTarget"] for r in results]
  average3 = sum(sizesOfTarget)/len(sizesOfTarget)

  print("Average size of Target network: ",average3)



  # |L|

  sizesOfL = [r["sizeOfL"] for r in results]
  average1 = sum(sizesOfL)/len(sizesOfL)

  print("Average size of learned network: ",average1)

  # Qa
  Qas = [r["QA"] for r in results]
  average4 = sum(Qas)/len(Qas)

  print("Average of Qas: ",average4)


  # Qc
  Qcs = [r["QC"] for r in results]
  average5 = sum(Qcs)/len(Qcs)

  print("Average of Qcs: ",average5)

  # nbWhatIsWrongQueriesA
  nbWhatIsWrongQueriesAs = [r["nbWhatIsWrongQueriesA"] for r in results]
  average20 = sum(nbWhatIsWrongQueriesAs)/len(nbWhatIsWrongQueriesAs)

  print("Average of nbWhatIsWrongQueriesAs: ",average20)


  # nbWhatIsWrongQueriesC
  nbWhatIsWrongQueriesCs = [r["nbWhatIsWrongQueriesC"] for r in results]
  average21 = sum(nbWhatIsWrongQueriesCs)/len(nbWhatIsWrongQueriesCs)

  print("Average of nbWhatIsWrongQueriesCs: ",average21)

  # nbConfirmationQueriesA
  nbConfirmationQueriesAs = [r["nbConfirmationQueriesA"] for r in results]
  average22 = sum(nbConfirmationQueriesAs)/len(nbConfirmationQueriesAs)

  print("Average of nbConfirmationQueriesAs: ",average22)


  # nbConfirmationQueriesC
  nbConfirmationQueriesCs = [r["nbConfirmationQueriesC"] for r in results]
  average23 = sum(nbConfirmationQueriesCs)/len(nbConfirmationQueriesCs)

  print("Average of nbConfirmationQueriesCs: ",average23)

  # nbReformulationQueriesA
  nbReformulationQueriesAs = [r["nbReformulationQueriesA"] for r in results]
  average24 = sum(nbReformulationQueriesAs)/len(nbReformulationQueriesAs)

  print("Average of nbReformulationQueriesAs: ",average24)


  # nbReformulationQueriesC
  nbReformulationQueriesCs = [r["nbReformulationQueriesC"] for r in results]
  average25 = sum(nbReformulationQueriesCs)/len(nbReformulationQueriesCs)

  print("Average of nbReformulationQueriesCs: ",average25)



  # Q bar
  averageExampleSizes = [r["averageExampleSize"] for r in results]
  average8 = sum(averageExampleSizes)/len(averageExampleSizes)

  print("Average of averageExampleSizes: ",average8)

  # tmax

  tmaxs = [r["tmax"] for r in results]
  average6 = sum(tmaxs)/len(tmaxs)

  print("Average of tmaxs: ",average6)


  # t bar
  ts = [r["t"] for r in results]
  average7 = sum(ts)/len(ts)

  print("Average of ts: ",average7)

  # timea
  timeAs = [r["timeA"] for r in results]
  average9 = sum(timeAs)/len(timeAs)

  print("Average of timeAs: ",average9)
  # timec
  timeCs = [r["timeC"] for r in results]

  average10 = sum(timeCs)/len(timeCs)

  print("Average of timeCs: ",average10)

  #  C
  runningtimelessthan1hour = [r["runningTime"] for r in results if r["runningTime"]<3600]
  print("Number of runs that took less than 1 hour: ", len(runningtimelessthan1hour))

  #'byQuAcq'
  byQuAcqs = [r["byQuAcq"] for r in results]
  average11 = sum(byQuAcqs)/len(byQuAcqs)

  print("Average of byQuAcqs: ",average11)


  # 'byNLP'
  byNLPs = [r["byNLP"] for r in results]
  average12 = sum(byNLPs)/len(byNLPs)

  print("Average of byNLPs: ",average12)

  #'TotalClassifyTimes'
  TotalClassifyTimess = [r["TotalClassifyTimes"] for r in results]
  average13 = sum(TotalClassifyTimess)/len(TotalClassifyTimess)

  print("Average of TotalClassifyTimess: ",average13)


  #'AverageClassifyTimes'
  AverageClassifyTimess = [r["AverageClassifyTimes"] for r in results]
  average14 = sum(AverageClassifyTimess)/len(AverageClassifyTimess)

  print("Average of AverageClassifyTimess: ",average14)



  #'TotalNERTimes'
  TotalNERTimess = [r["TotalNERTimes"] for r in results]
  average15 = sum(TotalNERTimess)/len(TotalNERTimess)

  print("Average of TotalNERTimess: ",average15)

  #'AverageNERTimes'
  AverageNERTimess = [r["AverageNERTimes"] for r in results]
  average16 = sum(AverageNERTimess)/len(AverageNERTimess)

  print("Average of AverageNERTimes: ",average16)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def NLPEffectOnTotalQueriesC(prob,results):
  plt.figure(figsize=(10,9))
  rrr = zip([r["byNLP"] for r in results], [r["QC"] + r["nbWhatIsWrongQueriesC"] + r["nbConfirmationQueriesC"] + r["nbReformulationQueriesC"]   for r in results])
  l = list(rrr)
  l.sort(key = lambda k: k[0])
  plt.plot([r[0] for r in l], [r[1] for r in l])
  plt.title(f"Total number of queries until convergence by Number of constraints aquired by NLP\n{prob}")

def NLPEffectOnTotalQueriesA(prob,results):
  plt.figure(figsize=(10,9))
  rrr = zip([r["byNLP"] for r in results], [r["QA"] + r["nbWhatIsWrongQueriesA"] + r["nbConfirmationQueriesA"] + r["nbReformulationQueriesA"]   for r in results])
  l = list(rrr)
  l.sort(key = lambda k: k[0])
  plt.plot([r[0] for r in l], [r[1] for r in l])
  plt.title(f"Total number of queries until learning an equivalent network by Number of constraints aquired by NLP\n{prob}")

```
