```python
path = "/home/med/cspdata/ToFocusOnNow/distances/fourthAttempt/"
```

```python
o = open(path+"admin.jsonl","r")
l = []
for e in list(o):
    l.append(json.loads(str(e)))
    
```

```python
f = open("Data4.txt","w")
f.write("-DOCSTART-"+"	_	_	"+"O\n\n")
for k in range(len(l)):

    spans=[]
    entities = []
    l[k]['label'] = sorted(l[k]["label"], key=lambda x:x[0])
    for e in l[k]["label"]:
        entities.append(l[k]["text"][e[0]:e[1]])
        spans.append((e[0],e[1],e[2]))

    i = 0
    j = -1
    while(True):
        j+=1
        tmp = spans[j]
        if(l[k]["text"][i:tmp[0]-1]!=""):
            print(l[k]["text"][i:tmp[0]-1] + "	_	_	" +"O")
            f.write(l[k]["text"][i:tmp[0]-1] + "	_	_	"+"O\n")
        print(l[k]["text"][tmp[0]:tmp[1]] + "	_	_	"+ tmp[2])
        f.write(l[k]["text"][tmp[0]:tmp[1]] + "	_	_	"+ tmp[2]+"\n")
        i=tmp[1]        
        if j == len(spans)-1:
            print(l[k]["text"][tmp[1]:] + "	_	_	"+ "O")
            f.write(l[k]["text"][tmp[1]:] + "	_	_	"+ "O\n")
            break    

    spans= []
    entities = []
    print("\n\n----------------------------------\n\n")
    f.write('\n')

f.close()
```
